```python
import pandas as pd
import altair as alt
import os # Importing the os module to interact with the file system

# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)
# print(os.listdir('/content/drive/MyDrive'))
# Read the CSV file into a DataFrame
df = pd.read_table('/content/drive/MyDrive/Track_Progress.csv', delimiter=',')
df.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert First Viewed and Last Viewed columns to datetime
df['First Viewed'] = pd.to_datetime(df['First Viewed'])
df['Last Viewed'] = pd.to_datetime(df['Last Viewed'])

# Calculate the difference between Last Viewed and First Viewed in hours
df['Time Spent'] = (df['Last Viewed'] - df['First Viewed']).dt.total_seconds() / 3600

# Group the data by Student Name and find the sum of Time Spent
df_agg = df.groupby('Student Name')['Time Spent'].sum().reset_index()

# Aggregate the data by Student Name and find the sum of Time Spent and Times Viewed
df_agg = df.groupby('Student Name').agg({'Time Spent': 'sum', 'Times Viewed': 'sum'}).reset_index()

# Rename the columns to Total Time Spent and Total Times Viewed
df_agg.columns = ['Student Name', 'Total Time Spent', 'Total Times Viewed']

# Sort the data in descending order by Total Time Spent
df_agg = df_agg.sort_values('Total Time Spent', ascending=False)

# Create a bar plot with Student Name on the x-axis and Total Time Spent on the y-axi
plt.figure(figsize=(10, 6))
# dispay x-axis labels vertically
plt.xticks(rotation=90)

#label axis
plt.xlabel('Student Name')
plt.ylabel('Total Time Spent')
plt.bar(df_agg['Student Name'], df_agg['Total Time Spent'])

# Save the chart
plt.savefig('total_time_spent_by_each_student_bar_chart.png')

# Create a bar plot with Student Name on the x-axis and Total Times Viewed on the y-axis
plt.figure(figsize=(10, 6))
# display x-axis values vertically
plt.xticks(rotation=90)
#label axis
plt.xlabel('Student Name')
plt.ylabel('Total Times Viewed')
plt.bar(df_agg['Student Name'], df_agg['Total Times Viewed'])

# Save the chart
plt.savefig('total_times_viewed_by_each_student_bar_chart.png')
plt.title('Total Time Spent by Each Student')
plt.title('Total Times Viewed by Each Student')

```

```python
# Display the average Total times Viewed by each student
print(df_agg['Total Times Viewed'].mean())
```

```python
# Plot a graph for every student who viewed for the total time at least 28 times, followed those who viewed less than 28 times
df_agg[df_agg['Total Times Viewed'] >= 28].plot(x='Student Name', y='Total Times Viewed', kind='bar')
df_agg[df_agg['Total Times Viewed'] < 28].plot(x='Student Name', y='Total Times Viewed', kind='bar')
```
